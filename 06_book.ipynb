```python
# example from the book about how negative correlations can arise.
nw = np.random.normal(0, 1, 100)
tw = np.random.normal(0, 1, 100)
plt.scatter(nw, tw)

s = nw + tw
mask = s >= np.quantile(s, 0.9)
plt.scatter(nw[mask], tw[mask])
p = np.corrcoef(nw[mask], tw[mask])[0, 1]
print(p)
plt.gca().set_aspect(1)
```
